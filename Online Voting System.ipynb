```python
# Online Voting System

# Dictionary to store voters and whether they have voted
voters = {
    "V001": False,
    "V002": False,
    "V003": False,
    "V004": False
}

# Candidate list
candidates = {
    1: {"name": "Alice", "votes": 0},
    2: {"name": "Bob", "votes": 0},
    3: {"name": "Charlie", "votes": 0}
}

def display_candidates():
    print("\n--- Candidate List ---")
    for num, candidate in candidates.items():
        print(f"{num}. {candidate['name']}")

def cast_vote(voter_id):
    if voters[voter_id]:
        print("‚ö†Ô∏è You have already voted! Duplicate voting is not allowed.")
        return
    display_candidates()
    try:
        choice = int(input("Enter the candidate number you want to vote for: "))
        if choice in candidates:
            candidates[choice]["votes"] += 1
            voters[voter_id] = True
            print(f"‚úÖ Vote cast successfully for {candidates[choice]['name']}")
        else:
            print("‚ùå Invalid choice. Please try again.")
    except ValueError:
        print("‚ùå Invalid input. Please enter a number.")

def display_results():
    print("\n--- Voting Results ---")
    for candidate in candidates.values():
        print(f"{candidate['name']}: {candidate['votes']} votes")

def vote_summary():
    total_votes = sum(candidate["votes"] for candidate in candidates.values())
    print("\n--- Vote Summary ---")
    print(f"Total votes cast: {total_votes}")
    for candidate in candidates.values():
        percentage = (candidate["votes"] / total_votes * 100) if total_votes > 0 else 0
        print(f"{candidate['name']}: {candidate['votes']} votes ({percentage:.2f}%)")

def main():
    while True:
        print("\n--- Online Voting System ---")
        print("1. Login and Vote")
        print("2. Display Results")
        print("3. Vote Summary")
        print("4. Exit")
        
        choice = input("Enter your choice: ")
        
        if choice == "1":
            voter_id = input("Enter your Voter ID: ")
            if voter_id in voters:
                cast_vote(voter_id)
            else:
                print("‚ùå Invalid Voter ID.")
        elif choice == "2":
            display_results()
        elif choice == "3":
            vote_summary()
        elif choice == "4":
            print("üëã Exiting... Thank you for participating!")
            break
        else:
            print("‚ùå Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
